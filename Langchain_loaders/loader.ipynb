# Directory Loader


```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader
```

```python
loader = DirectoryLoader(
    path = "books",
    glob="*pdf",
    loader_cls=PyPDFLoader,
    show_progress=True
)
```

```python
docs = loader.lazy_load()
```

```python
for document in docs:
    print(document.metadata)
```

# Web base loader


```python
from langchain_community.document_loaders import WebBaseLoader
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatGroq(
    model = "llama3-8b-8192",
    groq_api_key = "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"
)
```

```python
prompt = PromptTemplate(
    template='Answer the following question \n {question} from the following text - \n {text}',
    input_variables=['question','text']
)
```

```python
parser = StrOutputParser()
```

```python
url = "https://www.flipkart.com/6bo/b5g/~cs-8f246f75b2/pr?sid=6bo,b5g&collection-tab-name=Acer+Swift+Neo+OLED&pageCriteria=default&sort=price_asc"
loader = WebBaseLoader(url)
docs = loader.load()
```

```python
chain=prompt|llm|parser
```

```python
print(chain.invoke({'question':'What is the prodcut that we are talking about?', 'text':docs[0].page_content}))
```
